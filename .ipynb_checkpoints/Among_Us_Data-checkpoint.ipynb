```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'Jakub Questionnaire.csv'
data_df = pd.read_csv(path)
```

```python
plt.hist(data_df.colour, bins=5)
plt.xlabel('Colour choice')
plt.ylabel('Choice amount')
plt.title('Favourite primary colours')
```

```python
plt.hist(data_df.darkmode, bins=5)
plt.xlabel('Answer amount')
plt.ylabel('Viewing preference')
plt.title('Peoples software/browser viewing preference')
```

```python
plt.hist(data_df.glasses, bins=3)
plt.xlabel('Answer amount')
plt.title('Peoples dependancy on sight aid')
```

```python
plt.rcParams["figure.figsize"] = [8.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.hist(data_df.Breaks, bins=25)
plt.xlabel('Answer amount')
plt.ylabel('Break frequency')
plt.title('How often do people take breaks from their screens')
```

```python
plt.hist(data_df.bed, bins=7)
plt.xlabel('Answer amount')
plt.ylabel('Bed side wakeup choice')
plt.title('What side of the bed do people normally get out on')
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

path = 'Jakub Questionnaire.csv'
data_df = pd.read_csv(path)
correlation_table = pd.crosstab(data_df['glasses'], data_df['darkmode'])

sns.heatmap(correlation_table, annot=True, cmap='coolwarm', fmt='d')
plt.title('correlation of people who wear glasses and use darkmode')
plt.show()
```
