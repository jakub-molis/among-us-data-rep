```python
active_complaints_df = complaints_df[[complaint for complaint in complaints_df.Status == 'ACTIVE']]
```

```python
active_complaints_df
```

```python
leaky_boiler_df = complaints_df[[boiler for boiler in complaints_df['Complaint Category'] == '56']]
```

```python
house_streets = leaky_boiler_df['House Street'].tolist()
```

```python
house_streets_edit = [x.strip()for x in house_streets]
house_streets_edit
```

```python
house_num = leaky_boiler_df["House Number"].tolist()
```

```python
bad_places = [num + " " + street for num,street in zip(house_num, house_streets)]
```

```python
bad_places
```

```python
from collections import Counter

worst_places = Counter(bad_places).most_common(5)
print(worst_places)
```
